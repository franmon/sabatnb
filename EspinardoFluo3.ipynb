```python
def load_images(ipath):
    
    DIRS = {}
    # r=root, d=directories, f = files
    for r, drs, f in os.walk(ipath):
        print(len(f))
        if len(drs) > 1:
            for d in drs:
                dd = d.split('_')
                s1 = dd[2].replace(',0','')
                s2 = dd[-1].replace(';','-')
                lbl = s1+'_'+s2
                print(lbl)
                
                files = []
                for file in f:
                    if '.jpg' in file:
                        files.append(os.path.join(r, file))
                print(len(files))
                NPI ={}
                PDF ={}
                for ff in files:
                    f0  = ff.split('/')[-1]
                    ff0 = f0.split('_')
                    key = ff0[-1]
                    print(key)
        
                    im = Image.open(f)
                    npi = np.asarray(im)
                    df = pd.DataFrame(npi, index=range(npi.shape[0]))
                    NPI[key] = npi 
                    PDF[key] = df
                DIRS[lbl] = (NPI,PDF)
    return DIRS
```
